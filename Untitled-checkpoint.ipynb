## Data cleaning

```python
# import necessary libraries for whole project
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
```

First we download the dataset that one coursemate was able to acquire. <br>
Print info and first 5 rows of the data.

```python
# load data
df = pd.read_csv("../data/asunnot.csv")

# print info of data and first 5 rows
print(df.info())
print("\n")
print(df.head())
```

There are some categorial and numeric variables.
Some of the features are really important and some seems to be useless or very scrappy.<br>
I decided to go with 10 largest city in Finland by population.
I felt that model would be most accurate in the biggest cities.

```python
# Choose 10 largest cities from Kaupunki column
df = df.loc[df["Kaupunki"].isin(["Helsinki",
                                 "Espoo",
                                "Tampere",
                                "Vantaa",
                                "Oulu",
                                "Turku",
                                "Jyväskylä",
                                "Kuopio",
                                "Lahti",
                                "Pori"])]
# Print unique values of Cities
print(df.Kaupunki.unique())
```

Lets drop few of the columns straight away.<br>
* Huoneisto: Even though it tells us more information about the apartment it is really disorganized feature. In some apartments it includes a bit more info but in some apartments it doesen't offer us anything. Parsing through this feature would take way too much time to be woth the info available
* €/m2: This feature basically just tells us the selling price of the apartment. It would make the models too unreliable
* Kaupunginosa: This is pretty much the same as Postinumero (Zipcode) but less accurate

```python
# Drop columns Huoneisto, €/m2 and Kaupunginosa
df= df.drop(columns=["Huoneisto","€/m2","Kaupunginosa"])
```

Next we convert categorial variables to correct data type, check the unique values and print the amount of Nulls in them

```python
# Convert categorial variable to correct data type
df["Kaupunki"] = df["Kaupunki"].astype("category")
df["Huoneistotyyppi"] = df["Huoneistotyyppi"].astype("category")
df["Talotyyppi"] = df["Talotyyppi"].astype("category")
df["Hissi"] = df["Hissi"].astype("category")
df["Kunto"] = df["Kunto"].astype("category")
df["Tontti"] = df["Tontti"].astype("category")
df["Energialuokka"] = df["Energialuokka"].astype("category")
```

```python
# Check unique values of categorial variables (no cities since we know them already from earlier)
print("Huoneistotyyppi categories:",df.Huoneistotyyppi.unique(),"\n")
print("Talotyyppi categories:",df.Talotyyppi.unique(),"\n")
print("Hissi categories:",df.Hissi.unique(),"\n")
print("Kunto categories:",df.Kunto.unique(),"\n")
print("Tontti categories:",df.Tontti.unique(),"\n")
print("Energialuokka categories:",df.Energialuokka.unique(),"\n")

# Printing amount of Nulls in categorial variables
print("Amount of Nulls:")
print(df[["Huoneistotyyppi","Talotyyppi","Hissi","Kunto","Tontti","Energialuokka"]].isnull().sum())
```

We can see there is Nulls in three categorial features. Kunto tells us the apartments condition which has high correlation to the selling price. Kunto is really important feature thus we drop all rows where Kunto is Null. Tontti and Energialuokka are not as important so we can just fill those Nulls with 0.

```python
# Drop rows with NaN in Kunto column
df = df.dropna(subset ="Kunto")

# First set new category 0
# Fill all NaN from Tontti and Energialuokka with value 0
df["Tontti"] = df["Tontti"].cat.add_categories("0").fillna("0")
df["Energialuokka"] = df["Energialuokka"].cat.add_categories("0").fillna("0")

# Check amount of NaN in categorial variables, all should be 0
print(df[["Huoneistotyyppi","Talotyyppi","Hissi","Kunto","Tontti","Energialuokka"]].isnull().sum())
```

Next we do the same to the numeric values

```python
# First check amount of NaN in numeric variables
print(df[["Postinumero","m2","Velaton hinta","Rakennusvuosi","Kerros"]].isnull().sum())
```

Here we have the same sitsuation with Postinumero as we had earlier with Kunto. Postinumero is a really important feature since it tells use the more precises location. <br>
We also modify the Kerros feature a little and fill the Nulls with 0 since it is a less important feature.

```python
# Drop rows with NaN in Postinumero column
df = df.dropna(subset ="Postinumero")

# First covert Kerros to categorial temporarily
df["Kerros"] = df["Kerros"].astype("category")
# Fill all NaN in Kerros with value 0/0
df["Kerros"] = df["Kerros"].cat.add_categories("0/0").fillna("0/0")

# Then split Kerros column in two and drop original
# First one is the floor where the apartment is
# Second one is the amount of floors in the building
df[["Asunnon_kerros", "Kerrosten_lkm"]] = df.Kerros.str.split("/", expand = True)
df= df.drop(columns=["Kerros"])


# Now we can convert Asunnon_kerros and Kerrosten_lkm to numeric variables
df["Asunnon_kerros"] = df["Asunnon_kerros"].astype("int")
df["Kerrosten_lkm"] = df["Kerrosten_lkm"].astype("int")
```

Let's print the basic statistics of numeric values

```python
# Print description of numeric variables
print(df[["m2","Velaton hinta","Rakennusvuosi","Asunnon_kerros","Kerrosten_lkm"]].describe())
```

There are few things we have to fix here:
* There is atleast one apartment with living are of 1 square meter.
    * This is way too small to be an apartment
    * I decided to set the threshold at 10 square meters.
* There is apartment with selling price of 1
    * I decided to set the threshold to 1 and see which is the next minimum and see if this threshold needs modification
* There are datapoints with negative floors of the apartment or negative total amount of floors in the buildin
    * I set this to 0 (Handled them like Nulls)

```python
# Choose rows where m2 is more than 10
df= df[df.m2 > 10]

# Select rows where Velaton hinta is more than 1
df = df[df["Velaton hinta"] > 1]

# Replace negative Asunnon_kerros and Kerrosten_lkm values with 0
df.loc[df["Asunnon_kerros"]<0,"Asunnon_kerros"] = 0
df.loc[df["Kerrosten_lkm"]<0,"Kerrosten_lkm"] = 0


print(df[["m2","Velaton hinta","Rakennusvuosi","Asunnon_kerros","Kerrosten_lkm"]].describe())
```

Now the description looks good to continue.<br>
